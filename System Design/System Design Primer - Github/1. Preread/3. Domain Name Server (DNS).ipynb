# DNS

- DNS translates a recognisable english URL (https://google.com) into an IP address
- DNS servers store a map of URL to IP address, and everytime you key in a URL, your browser sends a query to the DNS server to get the requested address (assuming it isn't already in your browser cache)
    - Servers are queried in a hierachy, because some servers are more important than others
    - So you will first query the top level DNS server first (probably maintained by your ISP)
    - If the name resolution fails, then you will go down the hierachy to lower level DNS servers

- There are a few components for a given domain name on a DNS server
    - NS (name server) Record
        - Specifies the authoritative name server for a give domain, so you don't end up querying some random DNS down the hierachy
        - Suppose you want to access https://example.com
            - Query will route you to the root name server for the `.com` top level domain (TLD)
            - The TLD name server will provide the NS record for `https://example.com`, which points you to the authoritative name server for the domain
            - The authoritative name server responds with the IP address
    - MX record (mail exchange)
        - For a given domain, there is a specific mail server. This record directs email to the right mail server
        - Suppose you want to send an email to `user@example.com`
            - The sending mail server queries DNS for MX record of `example.com`
            - A single domain can have multiple MX records, and the email sending will be attempted in order of priority
            - The receiving mail server will be in charge of 
                - receiving the mail from the sending mail server
                - perform validation and security checks
                - verify address and check recipient mailbox space
                - email stored in recipient's mailbox on mail server via mail delivery agent (e.g. Dovecot)
                - Send notification
    - Address record
        - Map domain to IP address
    - CNAME (canonical)
        - Point to another name or CNAME, or to a record
    

- Managed DNS solutions
    - Cloudflare
    - Route53

- Routing management
    - Weighted round robin (avoid servers under maintainence)
    - Latency based (prioritise servers with lower latency)
    - Geolocation based (prioritise servers that are nearer by distance)

- Downsides
    - Accessing a DNS server introduces a slight delay, although mitigated by caching described above.
    - DNS server management could be complex and is generally managed by governments, ISPs, and large companies.
    - DNS services have recently come under DDoS attack, preventing users from accessing websites such as Twitter without knowing Twitter's IP address(es).
